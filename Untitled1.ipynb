```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np

# Fix missing SHX
os.environ["SHAPE_RESTORE_SHX"] = "YES"

# ------------------------------------
# Load India States Map
# ------------------------------------
india_states = gpd.read_file("gadm41_IND_1.shp")

# Filter Maharashtra
maharashtra = india_states[india_states["NAME_1"] == "Maharashtra"]

# ------------------------------------
# Jayakwadi Catchment Boundary
# ------------------------------------
lat_min = 18.5
lat_max = 20.2
lon_min = 73.5
lon_max = 76.5

boundary_lat = [
    lat_min, lat_min, lat_max, lat_max, lat_min
]

boundary_lon = [
    lon_min, lon_max, lon_max, lon_min, lon_min
]

# ------------------------------------
# Jayakwadi Dam Location
# ------------------------------------
dam_lat = 19.4950
dam_lon = 75.2280

# ------------------------------------
# Create 30 Rainfall Grid Points
# ------------------------------------
lat_range = np.arange(18.5, 20.21, 0.34)
lon_range = np.arange(73.5, 76.51, 0.75)

points = []

for lat in lat_range:
    for lon in lon_range:
        points.append([lat, lon])

points_df = pd.DataFrame(points, columns=["Latitude", "Longitude"])

print("Total Points:", len(points_df))

# ------------------------------------
# Plot
# ------------------------------------
fig, ax = plt.subplots(figsize=(12, 10))

# Plot Maharashtra Map
maharashtra.plot(
    ax=ax,
    color="lightgray",
    edgecolor="black"
)

# Plot Catchment Boundary
ax.plot(
    boundary_lon,
    boundary_lat,
    linewidth=2,
    marker="o",
    color="green",
    label="Catchment Area"
)

# Plot Dam
ax.scatter(
    dam_lon,
    dam_lat,
    marker="*",
    s=350,
    color="red",
    label="Jayakwadi Dam"
)

# Plot Rainfall Points
ax.scatter(
    points_df["Longitude"],
    points_df["Latitude"],
    color="blue",
    s=40,
    alpha=0.7,
    label="Rainfall Points"
)

# Label Dam
ax.text(
    dam_lon + 0.05,
    dam_lat + 0.05,
    "Jayakwadi Dam",
    fontsize=11,
    fontweight="bold"
)

# ------------------------------------
# Settings
# ------------------------------------
ax.set_title("Jayakwadi Dam Catchment with Rainfall Grid (Maharashtra)")

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

ax.legend()
ax.grid(True)

# ------------------------------------
# Show
# ------------------------------------
plt.show()
```

```python
import requests
import pandas as pd
import time
from tqdm import tqdm


# ----------------------------------
# LAT-LON POINTS (30)
# ----------------------------------

latitudes = [
    18.50, 18.84, 19.18, 19.52, 19.86, 20.20
]

longitudes = [
    73.50, 74.25, 75.00, 75.75, 76.50
]


# ----------------------------------
# YEARS RANGE (10 YEARS)
# ----------------------------------

START_YEAR = 2016
END_YEAR   = 2025     # inclusive


# ----------------------------------
# FETCH FUNCTION
# ----------------------------------

def fetch_rainfall(lat, lon, start_date, end_date):

    url = "https://archive-api.open-meteo.com/v1/archive"

    params = {
        "latitude": lat,
        "longitude": lon,
        "start_date": start_date,
        "end_date": end_date,
        "daily": "precipitation_sum",
        "timezone": "Asia/Kolkata"
    }

    try:
        response = requests.get(url, params=params, timeout=60)

        if response.status_code != 200:
            print("Failed:", lat, lon, start_date)
            return None

        return response.json()

    except Exception as e:
        print("Error:", e)
        return None


# ----------------------------------
# MAIN LOOP
# ----------------------------------

all_data = []

print("Fetching 10 Years Rainfall Data...")


for year in range(START_YEAR, END_YEAR + 1):

    start_date = f"{year}-01-01"
    end_date   = f"{year}-12-31"

    print(f"\nProcessing Year: {year}")


    for lat in tqdm(latitudes):

        for lon in longitudes:

            data = fetch_rainfall(lat, lon, start_date, end_date)

            if data is None:
                continue

            dates = data["daily"]["time"]
            rainfall = data["daily"]["precipitation_sum"]

            for d, r in zip(dates, rainfall):

                all_data.append([
                    d,
                    lat,
                    lon,
                    r
                ])

            # Anti-ban delay
            time.sleep(1)



# ----------------------------------
# CREATE DATAFRAME
# ----------------------------------

df = pd.DataFrame(
    all_data,
    columns=["Date", "Latitude", "Longitude", "Rainfall"]
)


# ----------------------------------
# SAVE
# ----------------------------------

df.to_csv("jayakwadi_rainfall_2016_2025(1).csv", index=False)


# ----------------------------------
# SUMMARY
# ----------------------------------

print("\nFinished!")
print("Total Rows:", len(df))
print(df.head())
```

```python

# ------------------------------------
# Load Raw Rainfall Data
# ------------------------------------
file_path = "jayakwadi_rainfall_2016_2025(1).csv"   # change name if needed

df = pd.read_csv(file_path)

print("Raw Data Shape:", df.shape)
print(df.head())


# ------------------------------------
# Convert Date Column
# ------------------------------------
df["Date"] = pd.to_datetime(df["Date"])


# ------------------------------------
# Take Daily Average of All Points
# ------------------------------------
daily_avg = (
    df
    .groupby("Date")["Rainfall"]
    .mean()
    .reset_index()
)

# Rename column (optional)
daily_avg.rename(columns={"Rainfall": "avg_rainfall"}, inplace=True)


# ------------------------------------
# Save Optimized File
# ------------------------------------
daily_avg.to_csv("jayakwadi_daily_avg_rainfall_2025.csv", index=False)


# ------------------------------------
# Show Result
# ------------------------------------
print("\nFinal Optimized Data:")
print(daily_avg.head())

print("\nTotal Days:", len(daily_avg))
```

```python
import pandas as pd


# ------------------------------------
# Load Daily Rainfall
# ------------------------------------
df = pd.read_csv("jayakwadi_daily_avg_rainfall_2025.csv")

print("Input Data:")
print(df.head())


# ------------------------------------
# Constants (Jayakwadi)
# ------------------------------------
CATCHMENT_AREA = 21750    # km
RUNOFF_COEFF   = 0.4


# ------------------------------------
# Calculate Inflow (m/day)
# ------------------------------------
df["Inflow (m3/day)"] = (
    df["avg_rainfall"] *
    CATCHMENT_AREA *
    1000 *
    RUNOFF_COEFF
)


# ------------------------------------
# Round Values
# ------------------------------------
df["Inflow (m3/day)"] = df["Inflow (m3/day)"].round(2)
df["avg_rainfall"] = df["avg_rainfall"].round(2)


# ------------------------------------
# Keep Required Columns
# ------------------------------------
final_df = df[["Date", "avg_rainfall", "Inflow (m3/day)"]]


# ------------------------------------
# Save
# ------------------------------------
final_df.to_csv(
    "jayakwadi_rainfall_final_2016_2025.csv",
    index=False
)



# ------------------------------------
# Show
# ------------------------------------
print("\nFinal Output:")
print(final_df.head())

# print("\nSaved as: jayakwadi_rainfall_inflow_2016_2025.csv")
```
